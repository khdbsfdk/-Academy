```python
#하나의 페이지에 여러 그래프 그리기
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2*np.pi*x1) * np.exp(-x1)
y2 = np.cos(2*np.pi*x2)

ax1 = plt.subplot(2,1,1)  # subplot(행의 수, 열의 수, 위치)
plt.plot(x1, y1, 'yo-')
plt.title('A table of two sunplot')
plt.ylabel('Damped oscillation')
print(ax1)

ax2 = plt.subplot(2,1,2)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time')
plt.ylabel('Undamped')
print(ax2)
plt.show()
```

```python
# 선 그래프
df = pd.read_excel('data/시도별 전출입 인구수.xlsx', header=0)
#전출지의 NaN 자료를 수정 -> 이전 자료의 값으로 대체
df.fillna(method='ffill', inplace=True) #빈 값을 채운다, 이전 자료들로
df.head(25)
#데이터 전처리 작업
```

```python
#서울에서 다른 지역으로 전출한 자료만 추출
mask = (df['전출지별'] == '서울특별시') & (df['전입지별'] != '서울특별시')
df_soul = df[mask]
print(df_soul.head())
print()
# '전출지별'은 같은 자료니까 삭제
df_soul = df_soul.drop('전출지별', axis=1,)
print(df_soul)
print()
#전입지별을 전입지로 변경
df_soul = df_soul.rename({'전입지별': '전입지'}, axis=1)
df_soul = df_soul.set_index('전입지')
df_soul
```

```python
#서울에서 경기도로 전입한 인구수를 그래프로 확인
sr_one = df_soul.loc['경기도'] #얜 시리즈 형태임
plt.figure(figsize=(15, 6)) # 그래프 사이즈 지정
plt.plot(sr_one.index, sr_one.values, marker='v', markersize=10, color='r')
plt.xticks(size=10, rotation='vertical') #x축의 데이터를 세로로 회전
plt.title("서울에서 경기도로 이전한 인구수", fontsize=20) #그래프 제목 지정
plt.xlabel('연도', size=20) #x축 이름
plt.ylabel('인구 수', size =20) #y축 이름
plt.legend(labels=['서울 -> 경기도'], loc='best', fontsize=13)
plt.show()
```

```python
#남북한 전력 잘전량
df = pd.read_excel('data/남북한발전전력량.xlsx')
df.fillna(method='ffill', inplace=True) #빈 값을 채운다, 이전 자료들로
df_south = df.loc[0:4]
df_north = df.loc[5:9]
df
```

```python
#일단 북한의 자료만 추려서 수력, 화력, 발전량을 분석
df_north.drop('전력량 (억㎾h)', axis='columns', inplace=True)
#발전 전력별 컬럼을 인덱스로 지정
df_north.set_index('발전 전력별', inplace=True)

#그래프 x가 인덱스가 되어야함
df_north = df_north.T
```

```python
#년도별로 증감율, 합계 컬럼명을 총발전량으로 변경
df_north = df_north.rename(columns={'합계':'총 발전량'})

#총 발전량 - 1년 -> 이전 년도의 총 발전량
df_north['총 발전량 - 1년'] = df_north['총 발전량'].shift(1)

#증감율 (총 발전량 / 이전 년도의 발전량 -1) * 100
df_north['증감율'] = ((df_north['총 발전량'] / df_north['총 발전량 - 1년'])-1)*100
df_north
```

```python

# 그래프 그리기 : 수력, 화력 -> 증감율
ax1 = df_north[['수력', '화력']].plot(kind='bar',figsize=(20,10), stacked=True)
ax2 = ax1.twinx()

ax2.plot(df_north.index, df_north['증감율'], ls='--', marker='o', color='red',
        label='전년대비 증감율(%)')
ax1.set_ylim(0,500)
ax2.set_ylim(-50,50)

ax1.set_xlabel('연도', size=20)
ax1.set_ylabel('발전량(억 Kwh)')
ax2.set_ylabel('전년 대비 증감율 (%)')

plt.title('북한 전력 발전량')
ax1.legend(loc='best')

plt.show()
```

```python
#남북한 전력 잘전량
df = pd.read_excel('data/남북한발전전력량.xlsx')
df.fillna(method='ffill', inplace=True) #빈 값을 채운다, 이전 자료들로
df_south = df.loc[0:4]
df_north = df.loc[5:9]
df
```

```python
# 남한의 자료만 추려서 수력, 화력, 발전량을 분석
df_south.drop('전력량 (억㎾h)', axis=1, inplace=True)
# 발전 전력별 컬럼을 인덱스로 지정
df_south.set_index('발전 전력별', inplace=True)

# 그래프 x가 인덱스가 되어야함
df_south = df_south.T
df_south
```

```python
#년도별로 증감율, 합계 컬럼명을 총발전량으로 변경
df_south = df_south.rename(columns={'합계':'해당 연도 총 발전량'})

#총 발전량 - 1년 -> 이전 년도의 총 발전량
df_south['이전 연도 총 발전량'] = df_south['해당 연도 총 발전량'].shift(1)

#증감율 (총 발전량 / 이전 년도의 발전량 -1) * 100
df_south['증감율'] = ((df_south['해당 연도 총 발전량'] / \
                    df_south['이전 연도 총 발전량'])-1)*100
df_south
```

```python
# 그래프 그리기 : 수력, 화력 -> 증감율
ax1 = df_south[['수력', '화력']].plot(kind='bar',figsize=(20,10), stacked=True)
ax2 = ax1.twinx()

ax2.plot(df_south.index, df_south['증감율'], ls='--', marker='o', color='red',
        label='전년대비 증감율(%)')
ax1.set_ylim(0,5000)
ax2.set_ylim(-30,30)

ax1.set_xlabel('연도', size=20)
ax1.set_ylabel('발전량(억 Kwh)', fontsize = 18)
ax2.set_ylabel('전년 대비 증감율 (%)', fontsize = 18)

plt.title('남한 전력 발전량', fontsize = 35)
ax1.legend(loc='best')

plt.show()
```

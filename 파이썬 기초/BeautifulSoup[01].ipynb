```python
movie_title = []
movie_point = []
idx =0

#영화 제목, 평점 검색

for line in soup.find_all('tr'):
    title = line.select_one('div.tit5')
#     print("title - ", title)
    if title:
        movie_title.append(title.get_text().strip('\n'))
#         print("movie_title 리스트 - ", movie_title)
        
    point = line.find('td', class_='point')
    if point:
        movie_point.append(point.get_text())
#         print("movie_point 리스트 - ", movie_point)
        
#순위, 영화 제목, 평점 출력
for idx, (title, point) in enumerate(zip(movie_title, movie_point)):
    print("순위: {}  영화 이름: {}  평점: {}".format(idx+1, title, point))
```

```python
class Student:
    def study(self):
        print("공부하세요")
class Teacher:
    def teacher(self):
        print("가르쳐주십쇼")
        
classroom  = [Student(), Teacher(), Student(),
              Student(), Teacher()]

for person in classroom:
    if isinstance(person, Student):
        #만약 학생 클래스에 해당하면 공부하세요 출력
        person.study()
    else:
        person.teacher()
```

```python
class Student:
    def __init__(self, name, kor, eng, math):
        self.name = name
        self.kor = kor
        self.eng = eng
        self.math = math
    def get_sum(self):
        return self,kor + self.eng + self.math
    def get_avg(self):
        return self.get_sum() / 3
    def __str__(self):
        return "{}\t{}\t{}".format(self.name, self.get_sum(), self.get_avg())
    def __eq__(self, value):
        return self.get_sum() == value.get_sum()
    def __ne__(self, value):
        return self.get_sum() != value.get_sum()
    def __gt__(self, value):
        return self.get_sum() > value.get_sum()
        
students = [
    Student("윤인성", 87, 98, 88),
    Student("연하진", 92, 98, 96),
    Student("구지연", 76, 96, 94),
    Student("나선주", 98, 92, 96),
    Student("윤아린", 95, 98, 98),
    Student("윤명월", 64, 88, 92)
]

print(type(students[0]))

student_a = Student("윤인성", 87, 98, 88)
student_b = Student("연하진", 92, 98, 96)
print("student_a == student_b =>", student_a == student_b)
print("student_a != student_b =>", student_a != student_b)

for student in students:
    print(str(student))
    if student == student_a:
        print(student.name)
    elif student == student_b:
        print(student.name)
```

```python
#클래스 변수
class Student:
    count = 0
    students = []
    def __init__(self, name, kor, eng, math):  #클래스 생성 부분
        self.name = name
        self.kor = kor
        self.eng = eng
        self.math = math
        
        Student.count += 1
        Student.students.append(self)
#         print("{}번째 자료 : {}".format(Student.count, self.name))
        
    @classmethod
    def print(cls):
        print("--------------------학생 목록---------------------")
        print("이름\t국어\t영어\t수학\t총점\t평균")
        # \t 문자열 사이에 탭 간격을 줄 때 사용
        for student in cls.students:
            print(str(student))
        print("----------------------------------------------------")
        
    def get_sum(self):
        return self.kor + self.eng + self.math
    def get_avg(self):
        return self.get_sum() / 3
    def __str__(self):
        return "{}\t{}\t{}\t{}\t{}\t{:0.3f}".format(self.name,
                            self.kor, self.eng, self.math, self.get_sum(), self.get_avg())
    #여기서 {:0.3f}는 소수점 3번째 자리까지 출력
    #__str__가 실질적인 출력 코드이며, str(인스턴스) 형태로 불러온다
        
# students = [
#     Student("윤인성", 87, 98, 88),
#     Student("연하진", 92, 98, 96),
#     Student("구지연", 76, 96, 94),
#     Student("나선주", 98, 92, 96),
#     Student("윤아린", 95, 98, 98),
#     Student("윤명월", 64, 88, 92)
# ]

# print()
# print("전체 인원 수 {}".format(Student.count))
```

```python
class Circle:
    def __init__(self, radius):
        self.__radius = radius   #private 변수
    def get_circum(self):
        return self.__radius*2*3.14
    def get_area(self):
        return self.__radius*self.__radius*3.14
    
    def get_radius(self): # 게터
        return self.__radius
    def set_radius(self, value): #세터
        self.__radius = value
```

```python
circle = Circle(10)
print(circle.get_area())

print(circle.get_radius())
print(circle.get_radius())
circle.set_radius(5)
print(circle.get_radius())
```

```python
# 학생들의 이름과 점수를 받아서 리스트에 저장항 후 자료 출력

class Student:
    count = 0
    students = []
    def __init__(self, name, kor, eng, math, output_list):  #클래스 생성 부분
        self.name = name
        self.kor = kor
        self.eng = eng
        self.math = math
        self.output_list = output_list
        
        Student.count += 1
        Student.students.append(self.output_list)
#         print("{}번째 자료 : {}".format(Student.count, self.name))
        
    @classmethod
    def print(cls):
        print("--------------------출력---------------------")
        for student in cls.students:
            print(str(student))
        print("----------------------------------------------")
        
    def get_sum(self):
        return self.kor + self.eng + self.math
    def get_avg(self):
        return self.get_sum() / 3
    def __str__(self):
        return "{}\t{}\t{}\t{}".format(self.name, self.kor, self.eng, self.math)
```

```python
output_list = []
while True:
    name = input("학생 이름 입력")
    if name == 'q' or name == 'ㅂ':
        break
    score = list(map(int, input("국어 영어 수학 점수 입력 >>>").split()))

    output_list.append(name)
    for i in score:
        output_list.append(i)
    print("output_list - ", output_list)
    Student(name, score[0], score[1], score[2], output_list)
    
Student.print()
```

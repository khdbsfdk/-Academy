```python
# 2. bins = [1,15,25,35,60,90]
#     labels=["미성년자", "청년", "중년", "장년", "노년"]
#     각 라벨별 승객수를 추출

#np.histogram으로 경계 값 리스트를 만듦(bins는 구간 나누는 수)
bins = [1,15,25,35,60,90]

count, bin_value = np.histogram(titanic['age'], bins) 
print(count, bin_value)

#구간의 이름을 부여
bin_names=["미성년자", "청년", "중년", "장년", "노년"]

#분석하고자 하는 자료에 bin_names 추가
titanic['연령대'] = pd.cut(x=titanic['age'],  #처리할 데이터
                          bins=bin_value,    #경계 값 리스트
                        labels=bin_names,   #구간의 이름 리스트
                         include_lowest='True')
titanic
```

```python
# 2. bins = [1,15,25,35,60,90]
#     labels=["미성년자", "청년", "중년", "장년", "노년"]
#     각 라벨별 승객수를 추출

# count, bin_value = np.histogram(df['horsepower'], bins=3)
count, bin_value = np.histogram(titanic['age'], bins=[1,15,25,35,60,90])

bin_names=["미성년자", "청년", "중년", "장년", "노년"]
for cnt, label in zip(count, bin_names):
    print(label, ':', cnt, "명", sep='', end='\t')
print()
print("count",count)
print("bin_value",bin_value)

#구간의 이름을 부여
bin_names=["미성년자", "청년", "중년", "장년", "노년"]

#분석하고자 하는 자료에 bin_names 추가
titanic['연령대'] = pd.cut(x=titanic['age'],  #처리할 데이터
                          bins=[1,15,25,35,60,90], 
                        labels=["미성년자", "청년", "중년", "장년", "노년"],  
                         include_lowest='True')
titanic
```

```python
#그룹 연산 데이터 변환 : group.transform(매핑 함수)

#그룹별 age 열의 평균 집계 연산
age_mean = class_group.age.mean()
print(age_mean)
print('\n')

#그룹별 age 열의 표준편차 집계 연산
age_std = class_group.age.std()
print(age_std)
print('\n')

#그룹 객체에 age 열을 iteration으로 z-score을 계산하여  출력
for key, group in class_group.age:
    group_zscore=(group - age_mean.loc[key])/age_std.loc[key]
    print(' -- origin : ', key)
    print(group_zscore.head(3))
    print()
```

```python
#z-score를 계산하는 사용자 함수 정의

def z_score(x):
    return (x - x.mean()) / x.std()

age_trans = class_group.age.transform(z_score)
print(age_trans.loc[[1,9,0]])

#위에 하던거 각 글래스 맨위 데이터만 뽑기
```

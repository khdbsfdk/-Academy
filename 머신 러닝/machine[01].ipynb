```python
import numpy as np
import pandas as pd
import sklearn
from sklearn.datasets import load_iris

iris = load_iris()
print('아이리스 타입 - ',type(iris))
print('key - ', iris.keys())
# iris

print()
print('data - \n', iris.data)
print('data type - ', type(iris.data))
```

```python
print('target - ', iris.target)
print('target type - ', type(iris.target))
print()
print('target_names - ', iris.target_names)
print('target_names type - ', type(iris.target_names))

#여기서 setosa는 0, versicolor는 1, virginica는 2로 인코딩 해놓은게 target임.
print()
print('feature_names - ', iris.feature_names)
print('feature_names type - ', type(iris.feature_names))
```

```python
#feature, target을 이용해서 DF 만들기
iris_df = pd.DataFrame(data = iris.data,
                       columns=iris.feature_names)
iris_df['target'] = iris.target
iris_df
```

```python
""" 지도학습의 분류
step 01. 데이터를 분리(training data, test data).
step 02. 학습 데이터를 기반으로 ML 알고리즘 적용하여 학습 모델 생성
step 03. test 데이터를 기반으로 분류 예측을 수행
step 04. 모델에 대한 성능 평가
"""
```

```python
#데이터 분류 함수
from sklearn.model_selection import train_test_split
from sklearn.tree            import DecisionTreeClassifier
from sklearn.metrics         import accuracy_score
iris_df
```

```python
#데이터 분류
#train_test_split 함수는 리턴 값이 피처와 타켓 등 다양하기 때문에 이렇게 호출한다.
# test_size를 0.2로 두면 20%는 테스트 용으로 둔다. (셔플도 필요함)
X_train, X_test, y_train, y_test = train_test_split(iris.data,
                                                   iris.target,
                                                   test_size = 0.2,
                                                   shuffle = True,
                                                   random_state = 100)

#train_test_split(사용할 데이터, 정답, 크기, 편향도, 매번 실행 시 같은 데이터 셋 적용)
```

```python
#학습 데이터는 120개 컬럼은 4개 -> (120, 4)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
print('train data - ', X_train)
print('train target - ', y_train)
print('test data - ', X_test)
print('test target - ', y_test)
```

```python
#알고리즘 모델 생성 - 학습시 fit() 사용

iris_dtc_model = DecisionTreeClassifier() #모델 생성
iris_dtc_model.fit(X_train, y_train) # 학습
```

```python
#분류 예측 - predict()
# predict 함수를 통해 피처를 넣으면 예측 결과를 리턴해줌
y_pred = iris_dtc_model.predict(X_test)
print('y_test (정답) -> ', y_test)
print('y_pred (예측) -> ', y_pred)

#shuffle = False면 정답이 2뿐이였지만 True로 바꾸면
```

```python
#예측 정확도 확인 -> shuffle가 들어가면 확률 올라감
print('acc - ', accuracy_score(y_test, y_pred))
```

```python
#혼공머신 교제 243p 실습 예제 실행

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

wine = pd.read_csv('https://bit.ly/wine_csv_data')

data = wine[['alcohol', 'sugar', 'pH']].to_numpy()
target = wine['class'].to_numpy()

#훈련 세트와 테스트 세트를 나눕니다.
train_input, test_input, train_target, test_target = train_test_split(data,
                                                                     target,
                                                                     test_size=0.2,
                                                                     random_state=42)
#다시 훈련세트에서 가져와 훈련세트와 검증 세트로 나눕니다.
sub_input, val_input, sub_target, val_target = train_test_split(train_input,
                                                                 train_target,
                                                                 test_size=0.2,
                                                                 random_state=42)
# print(sub_input.shape, val_input.shape) #잘하고있는지 확인

#중간 검증 실행 - sub_input, val_input, sub_target, val_target로 모델 생성
dt = DecisionTreeClassifier(random_state=42)
dt.fit(sub_input, sub_target)
print("최종 훈련 세트 점수", dt.score(sub_input, sub_target))
print("최종 검증 세트 점수", dt.score(val_input, val_target))

#교차검증 함수 적용
from sklearn.model_selection import cross_validate
cross_score = cross_validate(dt, train_input, train_target)
import numpy as np
print("교차 검증 확률 평균", np.mean(cross_score['test_score']))

#만약 훈련 세트를 섞은 후 10-폴트 교차 검증을 수행하려면?
splitter = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
score = cross_validate(dt, train_input, train_target, cv=splitter)
print("평균 정확도 : ", np.mean(score['test_score']))
```

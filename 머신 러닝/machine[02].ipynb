- 트리계열의 알고리즘 중에서 성능이 떨어지는 단점이 있을 수 있다 - 숫자의 특성이 반영되기 때문.

```python
iris = load_iris()
print(type(iris))
print(iris.keys,type(iris.keys()))

X_train, X_test, y_train, y_test = train_test_split(iris.data,
                                                   iris.target,
                                                   test_size = 0.2,
                                                   shuffle = False,
                                                   random_state = 200)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
X_train
```

```python
#불순도(impurity) - 해당 범주안에 서로 다른 데이터가 얼마나 섞여있는지
#불순도를 수치화 한 것 - 엔트로피(entropy) 즉, 불순도를 수치적으로 나타낸 척도
#지니(gini)도 엔트로피와 마찬가지

gscv_dtc_model = DecisionTreeClassifier()
param = {
        'criterion':['gini', 'entropy'],
        'max_depth': [1,2,3],  #깊이 - 트리의 노드 갯수
        'min_samples_split': [2, 3], #스플릿 갯수
        'splitter':['random', 'best']
}

```

```python
#GridSearchCV를 이용한 최적의 모델을 만드는 방법
grid_tree = GridSearchCV(gscv_dtc_model,
                        param_grid = param,
                        cv = 5,
                        refit = True)
#refit은 최적의 모델을 찾았을 경우 모델을 가지고 재학습을 하겠다는 의미
```

```python
grid_tree.fit(X_train, y_train) #학습
print(type(grid_tree.cv_results_))
grid_tree.cv_results_
```

```python
#grid_tree.cv_results_가 딕셔너리 이므로 DF 만들기 가능
score_frm = pd.DataFrame(grid_tree.cv_results_)
score_frm
```

```python
score_frm[['params', 'rank_test_score']]
```

```python
print("최적의 파라미터 확인")
print(grid_tree.best_params_)
print()
print("최적의 파라미터의 정확도")
print(grid_tree.best_score_)
print()
print()
print(grid_tree.best_estimator_)
print(grid_tree.best_index_)
```

```python
#수능
estimator = grid_tree.best_estimator_
y_pred = estimator.predict(X_test)
print("정답 - \n", y_test)
print()
print("guess - \n", y_pred)
print()
print("정확도 - \n", accuracy_score(y_test, y_pred))
```

### 이상치 값 제거
- 아예 제거하는 방법
- 이상치 값 변환 -> NaN을 변환 후 대체 값으로 처리

```python
q1 = tips['tip'].quantile(0.25)   #이 값을 1사분위 값으로 지정
q3 = tips['tip'].quantile(0.75)   #이 값을 3사분위 값으로 지정

#IOR - 1사분위 ~ 3사분위
iqr = 1.5*(q3-q1)
```

```python
def removeQutlier(frm, column):
    q1 = frm[column].quantile(0.25)   #이 값을 1사분위 값으로 지정
    q3 = frm[column].quantile(0.75)   #이 값을 3사분위 값으로 지정
    print(q1, q3) 
    
    tmp_frm = frm
    #IOR - 1사분위 ~ 3사분위
    iqr = 1.5*(q3-q1)
    print("iqr:", iqr)
    tmp_frm[column][ (tmp_frm[column] > (q3 + iqr)) | (tmp_frm[column] < (q3 - iqr))] = None
    #위 값은 결측치
    
    return tmp_frm
```

```python
missing_frm = removeQutlier(tips, 'total_bill')
missing_frm.boxplot(column='total_bill')
plt.show()
plt.close()
```

# 정규식 실습

```python
import re

result = re.match('(\d{2})-(\d{3,4})-(\d{4})', '02-123-1234')
print(result.groups())
print(result.group())
print(result.group(0))
print(result.group(1))
print(result.group(2))
print(result.group(3))
```

```python
import re

# 1
# 숫자 찾기
example = '2022 03 15 tuesday'
print(re.findall('\d', example))

# 숫자가 아닌 것 찾기
print(re.findall('\D', example))

p = re.compile('^[a-z]')
result = re.findall(p, example)
```

```python
# 2

# 이메일 주소를 찾기
example = '제 이메일은 hreeee@yonsei.ac.kr 입니다. hreeee@gmail.com으로 변경되었습니다.'

result = re.findall('[a-zA-Z]+@[a-zA-Z0-9-]+\.[a-zA-Z0-9-.]+[a-z]', example)
result
```

```python
# 3

# URL 찾기
example = 'https://www.google.com/에 네이버 검색하시면 https://www.naver.com/이 상단에 노출됩니다.'

# https://cholol.tistory.com/457
result = re.findall('http[s]?://(?:[a-zA-Z]|[0-9]|[$\-@\.&+:/?=]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', example)
result
```

```python
# 4

# 파일 찾기
example = '제 바탕화면 보시면 sonata01.png 파일과 avante02.jpg 파일 그리고 socar2021.gif 파일이 존재합니다.'

p = re.compile('[a-z] + .(png)', example)
result = re.findall(p, example)
result
```

- 강사님 답안

```python
import re

# 1
# 숫자 찾기
example = '2022 03 15 tuesday'
print(re.findall('[0-9]', example))

# 숫자가 아닌 것 찾기
print(re.findall('[^0-9]', example))
```

```python
# 2

# 이메일 주소를 찾기
example = '제 이메일은 hreeee@yonsei.ac.kr 입니다. hreeee@gmail.com으로 변경되었습니다.'

result = re.findall('[a-z]+@[a-z]+.[a-z]+.[a-z]+', example)
result
```

```python
# 3

# URL 찾기
example = 'https://www.google.com/에 네이버 검색하시면 https://www.naver.com/이 상단에 노출됩니다.'
print(re.findall('https://[a-z]{3}.[a-z]+.[a-z]+/', example))
```

```python
# 4

# 파일 찾기
example = '제 바탕화면 보시면 sonata01.png 파일과 avante02.jpg 파일 그리고 socar2021.gif 파일이 존재합니다.'

print(re.findall('[a-z0-9]+\.[(png)|(jpg)|(gif)]+', example))
```

# 노이즈 제거

```python
def pad_punc(sentence, punc):
    for p in punc:
        sentence = sentence.replace(p, " " + p + " ")

    return sentence

sentence = "Hi, my name is john!!"
print(pad_punc(sentence, [".", "?", "!", ","]))
```

```python
copers = \
'''
TOLD HER SHOWS A MARKED ADVANCE(?) IN THE DEVELOPMENT OF HER INTELLECT, AND IS AN INVALUABLE STIMULUS TO THE ACQUISITION OF LANGUAGE(!). I ASK ALL HER FRIENDS TO ENCOURAGE HER TO TELL THEM OF HER DOINGS, AND TO MANIFEST AS MUCH CURIOSITY AND PLEASURE IN HER LITTLE ADVENTURES AS THEY POSSIBLY CAN.
'''

# https://www.gutenberg.org/files/2397/2397-h/2397-h.htm
```

```python
def cleaning_text(text, punc, regex):

    text = text.lower()
    text = re.sub(regex, " ", text)
    for p in punc:
        text = text.replace(p, " " + p + " ")

    return text
```

```python
print(cleaning_text(copers, [".", "?", "!", ","], "[^a-zA-Z.,?!\n]"))
```

```python
from konlpy.tag import Okt
from konlpy.tag import Kkma

okt = Okt()
kkma = Kkma()

print('OKT 형태소 분석 :',okt.morphs("열심히 코딩한 당신, 연휴에는 여행을 가봐요"))
print('OKT 품사 태깅 :',okt.pos("열심히 코딩한 당신, 연휴에는 여행을 가봐요"))
print('OKT 명사 추출 :',okt.nouns("열심히 코딩한 당신, 연휴에는 여행을 가봐요"))
```

```python
import re
text = "I was wondering if anyone out there could enlighten me on this car."

# 길이가 1~2인 단어들을 정규 표현식을 이용하여 삭제
shortword = re.compile(r'\W*\b\w{1,2}\b')
print(shortword.sub('', text))
```

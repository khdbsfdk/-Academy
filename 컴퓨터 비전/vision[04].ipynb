### 키보드 이벤트 처리

```python
import numpy as np
import cv2
import sys

img = cv2.imread('Ch_2_Opencv_basic/fig/puppy.bmp', 0)
img_ori = img.copy()

while True:
    key = cv2.waitKey()
    if key == 27:
        print("사용자의 의해 종료")
        break
    elif key == ord('o'):
        img = cv2.imread('Ch_2_Opencv_basic/fig/puppy.bmp', 1)
        img1 = img_ori.copy()
        cv2.imshow('img', img)
        
    elif key == ord('e'):
        #엣지 처리
        img = cv2.Canny(img, 50, 150)
        cv2.imshow('img', img)
        
    elif key == ord('i'):
        #반전 효과
        img = 255 - img
        cv2.imshow('img', img)
        
    elif key == ord('s'):
        #저장
        cv2.imwrite('input_key.png', img)
        print('저장 완료')
        
    else:
        cv2.imshow('img', img)
cv2.destroyAllWindows()
```

## 마우스 이벤트 처리

```python
# cv2.setMouseCallback(windowName, onMouse, param = None) -> None
    # windowName: 마우스이벤트를 수행할 창 이름
    # onMouse: 마우스 이벤트 콜벡함수
    # param: 콜백함수에 전달할 데이터

# onMouse(event, x, y, flags, param) -> None
# event: 마우스 이벤트 종류 e.g., cv2.EVENT_LBUTTONDOWN
# x, y : 창을 기준으로 이벤트 발생좌표
# flags: 이벤트시 발생 상태 e.g., "ctrl"
# param: cv2.setMouseCallback()함수에서 설정한 데이터

# 함수 정의
def call_mouse(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDOWN: #좌클릭 시
        print('좌표 값 : X: {}, Y: {}'.format(x, y))
    elif event == cv2.EVENT_LBUTTONUP: #좌 클릭 떼면
        print("뗐다")
    elif event == cv2.EVENT_MOUSEMOVE: #마우스 움직일 때마다
        if flags == cv2.EVENT_FLAG_LBUTTON: 
            print('좌표 값 : X: {}, Y: {}'.format(x, y))
```

```python
import numpy as np
import cv2
import sys

img = np.ones((480, 640, 3), np.uint8)*255

cv2.namedWindow('image')
#마우스 콜백 함수
cv2.setMouseCallback('image', call_mouse, img)
cv2.imshow('image', img)


cv2.waitKey()
cv2.destroyAllWindows()
```

## Trackbar

- createTrackbar(trackbarName, windowName, value, count, onChange) -> None
- trackbarName: 트랙바 이름
- windowName : 트랙바를 생성할 창 이름
- value : 트랙바 위치 초기값
- count : 트랙바 최댓값, 최솟값은 0
- onChange :callback 함수 e.g., onChange(pos) 위치를 정수형태로 전달

```python
import numpy as np
import cv2
import sys

def call_track(pos):
    global mask, img
    mask_val = pos*mask
    cv2.imshow('image', mask_val)

# img =  np.zeros((480, 640, 3), np.uint8)
img_alpha = cv2.imread('Ch_2_Opencv_basic/fig/imgbin_sunglasses_1.png',
                      cv2.IMREAD_UNCHANGED)
mask = img_alpha[:, :, -1]    # mask는 알파 채널로 만든 마스크 영상
img = img_alpha[:, :, -1]

mask[mask > 0] = 1

cv2.namedWindow('image')
cv2.createTrackbar('level', 'image', 50, 255, call_track)
cv2.imshow('image', mask)

cv2.waitKey()
cv2.destroyAllWindows()
```

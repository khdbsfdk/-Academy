사진 복사

```python
import numpy as np
import cv2
import sys

img = cv2.imread("Ch_1_Opencv_Intro/fig/puppy.bmp")

img1 = img
img2 = img.copy()
img1[100:200, 200:300] = (0, 0, 225)
#주소를 복사하므로 변경 사항도 복사됨
# 따라서 copy() 사용

cv2.imshow('img', img)
cv2.imshow('img1', img1)
cv2.imshow('img2', img2)
# cv2.imshow('img4', img4)

cv2.waitKey()
cv2.destroyAllWindows()
```

마우스 좌표

```python
import numpy as np
import cv2
import sys

# 마우스 이벤트 콜백함수 정의
def mouse_callback(event, x, y, flags, param): 
    print("마우스 이벤트 발생, x:", x ," y:", y) # 이벤트 발생한 마우스 위치 출력

img = cv2.imread('Ch_2_Opencv_basic/fig/rudolf.png', 1)

cv2.namedWindow('image')  #마우스 이벤트 영역 윈도우 생성

cv2.setMouseCallback('image', mouse_callback)

while(True):

    cv2.imshow('image', img)

    k = cv2.waitKey(1) & 0xFF
    if k == 27:    # ESC 키 눌러졌을 경우 종료
        print("ESC 키 눌러짐")
        break
cv2.destroyAllWindows()
```

```python
import numpy as np
import cv2
import sys

# 준비물
src = cv2.imread('Ch_2_Opencv_basic/fig/airplane.bmp')
mask = cv2.imread('Ch_2_Opencv_basic/fig/mask_plane.bmp', 0)
dst = cv2.imread('Ch_2_Opencv_basic/fig/field.bmp')

# 비행기 붙이기
# cv2.copyTo(src, mask, dst)
dst[mask > 0] = src[mask > 0]
dst[mask > 0] = src[0, 0, 0]

cv2.imshow('src', src)
cv2.imshow('mask', mask)
cv2.imshow('dst', dst)

cv2.waitKey()
cv2.destroyAllWindows()
```
